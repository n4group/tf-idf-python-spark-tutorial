```python
import pyspark
import json

sc = pyspark.SparkContext('local[*]')

wikiFile = sc.textFile('wikidata-20170306-all.json.gz')

sample = wikiFile \
  .filter(lambda x: '{' in x) \
  .map(lambda x: json.loads(x[:-1])) \
  .take(1)

formatted = json.dumps(sample, sort_keys=True, indent=4, separators=(',', ': '))
print(formatted)

sc.stop()
```
